# Aproximación polinomial

```python
#Libriries
import numpy as np
import matplotlib.pyplot as plt
```

## Defino los vectores 'x' y 'y' (puntos en el plano)

```python
#lOS VECTORES DEBEN SER DE TIPO ARRAY, POR LO CUAL CONVIERTO LAS LISTAS EN ARRAY
x=np.array(list(range(1,30)))
print('El vector x')
print(x)

y_list=[107897, 174946.333333333, 177962, 180250.5, 160520.333333333, 
141436.5, 131626, 133203.5, 129063.833333333, 151999.5, 188374.333333333, 
188616.666666667, 195182.333333333, 187230, 78318.8333333333, 
168473.833333333, 132007.833333333, 178469.833333333, 141593.166666667, 
60709.3333333333, 134688.5, 160448.166666667, 201055.166666667, 
188938.166666667, 207163, 141887, 181458, 164167, 58621.4166666667]

y=np.array(y_list)

print('El vector y:')
print(y)

#La LONGITUD DE LOS VECTORES DEBE COINCIDIR
print('Longitud de x:')
print(len(x))

print('Longitud de y:')
print(len(y))
```

## Aproximación polinomial

```python
#Usar funcion polyfit, de la paquetería numpy
curve=np.polyfit(x,y,9) #toda 'x' vs 'y' y calcula el polinomio de grado n
print('Aproximacion de la curva esta dada por el polinomio:')
print(curve)

p=np.poly1d(curve)   #Define mi polinomio como una función p
```

```python
xp = np.linspace(0, 30, 100)
plt.plot(x, y, '.', xp, p(xp), '-')
```
