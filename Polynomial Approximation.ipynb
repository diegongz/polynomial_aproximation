```python
#APROXIMACION DE FUNCIONES
import numpy as np
import matplotlib.pyplot as plt

xl=list(range(1,20))
x=np.array(xl)

y2=list(range(10,0,-1))
yl=list(range(1,10))+y2
y=np.array(yl)


print('x:')
print(x)

print('y:')
print(y)

#aproximacion de la curva
curve=np.polyfit(x,y,2)
print('Aproximacion de la curva esta dada por el polinomio:')
print(curve)

p=np.poly1d(curve)

```

```python
xp = np.linspace(0, 20, 100)
plt.plot(x, y, '.', xp, p(xp), '-')
```
